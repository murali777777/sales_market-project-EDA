# 8. which product line is purchased in the highest quantity?

```python
highest_quantity=sales.groupby(["Product line", 'Quantity']).size().reset_index(name='count').sort_values(by='count', ascending=False)[0:5]
highest_quantity
```

```python
plt.figure(figsize=(4, 8))
plt.pie(highest_quantity['count'], labels=highest_quantity['Product line'] + ' - ' + highest_quantity['Quantity'].astype(str), autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Product Line and Quantity Combinations')
plt.show()
```

# 9. display daily sales by day of the week 

```python
#to create a dictionary set  in the day of weeks 
dw_mapping={
    0:"mon",1:"Tue",2:'Wed',3:"Thur",4:"Fri",5:"Sat",6:"Sun"
}
```

```python
#display daily sales by day of the week
sales["Date"].dt.dayofweek.map(dw_mapping)
```

# to create a new column in the data set:(to re-create in the data set )

```python
sales["day_of_week"]=sales["Date"].dt.dayofweek.map(dw_mapping)
```

```python
sales.head()
```

# we can bar plot of the data

```python
sales["day_of_week"].value_counts().plot(kind="bar")
```

# 10.what will the highest months of the sales

```python
month_mapping = {
    1: "Jan",
    2: "Feb",
    3: "March",
    4: "April",
    5: "May",
    6: "June"
}
```

```python
#what will the highest months of the sales
sales["month"]=sales["Date"].dt.month.map(month_mapping)
```
